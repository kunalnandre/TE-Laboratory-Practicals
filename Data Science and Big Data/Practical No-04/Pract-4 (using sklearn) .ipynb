```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
data= pd.DataFrame(data=boston.data, columns=boston.feature_names)
data
```

```python
# merged prices column with data frame

data['prices']=boston.target
data
```

```python
# Sagrigate "Prices" column and remaining colums in X and Y

x=data.drop(["prices"],axis=1)

y=data['prices']
```

```python
# import library.class and function 

from sklearn.model_selection import train_test_split as tts
```

```python
# convert into variables

x_train, x_test, y_train, y_test=tts(x, y, test_size=0.9)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
LR= LinearRegression()
```

```python
LR.fit(x_train, y_train)
```

```python
y_pred=LR.predict(x_test)
y_pred
```

```python
LR.score(x_test, y_pred)
```
